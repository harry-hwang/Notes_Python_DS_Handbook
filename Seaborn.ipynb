# Seaborn

plots linear regression automatically

Easy to deal with Dataframes


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
a = sns.load_dataset('flights')
```

```python
a.head()
```

## Scatter Plot

```python
sns.relplot(x = "passengers", y = "month", data = a)
```

```python
sns.relplot(x = "passengers", y = "month", hue = "year", data = a)
```

## Line Plot

```python
b = sns.load_dataset("tips")
```

```python
b.head()
```

```python
sns.relplot(x = "time", y = "tip", data = b, kind = "line")
```

## Plot categorical variable

```python
sns.catplot(x = "day", y = "total_bill", data = b)
```

```python
sns.catplot(x = "day", y = "total_bill", data = b, kind = "violin")
```

```python
sns.catplot(x = "day", y = "total_bill", data = b, kind = "boxen")
```

## Multi-Plot Grids

```python
a = sns.load_dataset("iris")
a.head()
```

```python
b = sns.FacetGrid(a, col = "species") # col可以替換成row, 基本上都是用catgorical變數
b.map(plt.hist, "sepal_length")
```

```python
a = sns.load_dataset("flights")
```

```python
b = sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style = "dark") # 用style來調整background
b= sns.PairGrid(a)
b.map(plt.scatter)
```

```python
sns.set(style = "white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y = "total_bill", data = a)
```

```python
sns.set(style = "white", color_codes=True)
a = sns.load_dataset("tips")
sns.boxplot(x="day", y = "total_bill", data = a)
sns.despine(offset=10, trim = True)
```
